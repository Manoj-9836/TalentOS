# TalentOS — Phase 11 Evaluation

This notebook drives `backend/scripts/run_evaluation.py`, which produces `docs/reports/metrics_report.md`.

It computes:

1. **Retrieval quality** — recall@K of the FAISS index against the reference top-50.
2. **Ranking quality** — Pearson + Spearman correlation between `candidate_rank_score` and `recruiter_response_rate`.
3. **Recruitability model** — surfaces training-time MAE / RMSE / R² and feature importances.

Run the cells in order, or just execute the script directly:

```
python -m backend.scripts.run_evaluation
```

````python
import subprocess, sys
from pathlib import Path

ROOT = Path.cwd()
print("Project root:", ROOT)
````

````python
result = subprocess.run(
    [sys.executable, "-m", "backend.scripts.run_evaluation"],
    cwd=str(ROOT),
    capture_output=True,
    text=True,
)
print(result.stdout)
if result.returncode != 0:
    print("STDERR:", result.stderr)
    raise RuntimeError(f"Evaluation failed (exit code {result.returncode})")
````

````python
from IPython.display import Markdown, display

report = (ROOT / "docs" / "reports" / "metrics_report.md").read_text(encoding="utf-8")
display(Markdown(report))
````
